```python
#3.1(c) To create an element-wise comparison(greater, graeter equal,less,lesss equal, equal, equal withih a tolerance)
#greater
a=np.array([10,20,30])
b=np.array([15,20,25])
print("Greater:",np.greater(a,b))
print("Greater Equal:",np.greater_equal(a,b))
print("Less:",np.less(a,b))
print("Less Equal:",np.less_equal(a,b))
print("Equal:",np.equal(a,b))

#Check if array are almost equal(within a tolerance)
c=np.array([1.0001,2.0001])
d=np.array([1.0002,2.0002])
print("Allclose:",np.allclose(c,d))   #True(within default tolerance)
```

```python
#3.2(b)Find indices of max and min numbers along a given axis
arr2d=np.array([[10,20,30],[5,25,40]])

#Index of max,min along axis 0 (column-wise)
print("Argmax(axis=0):",np.argmax(arr2d,axis=0))
print("Argmin(axis=0):",np.argmin(arr2d,axis=0))

#Index of max,min along axis 1 (row-wise)
print("Argmax(axis=1):",np.argmax(arr2d,axis=1))
print("Argmin(axis=1):",np.argmin(arr2d,axis=1))

arr3d=np.array([[10,20,30],[5,25,40],[30,5,15]])
#Index of max,min along axis 0 (column-wise)
print("Argmax(axis=0):",np.argmax(arr3d,axis=0))
print("Argmin(axis=0):",np.argmin(arr3d,axis=0))

#Index of max,min along axis 1 (row-wise)
print("Argmax(axis=1):",np.argmax(arr3d,axis=1))
print("Argmin(axis=1):",np.argmin(arr3d,axis=1))
```
